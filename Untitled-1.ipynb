```python

import matplotlib.pyplot as plt
import numpy as np
```

```python
def logistic( x0, r):
    x0 = np.float64(x0)
    r = np.float64(r)
    x1 = r * x0*(1 -x0)
    return x1

def iter(n,x0,r):
    xb = x0
    hist = []
    hist.append(x0)
    for i in range(n):
        x1 = logistic(xb,r)
        xb = x1
        hist.append(xb)
    return hist
```

```python
def bifurcation_plot(x0, r_a, r_b, step, n, k ):
    bif =[]
    r_range = np.arange(r_a, r_b, step)
    for r in r_range:
        hist = iter(n, x0, r)
    
        hist = hist[k-1:]
        bif.append(hist)

    return bif




```

```python
r_a = 2.8
r_b = 3.86
step = 0.0001
se = bifurcation_plot(0.2,r_a,r_b,step, 1500, 1400)
r_range = np.arange(r_a, r_b, step)
plt.plot(r_range,se,',')
#plt.xlim(3.84, 3.8571)
#plt.ylim(0.4,0.6)
plt.show()
```

```python
def get_b(bif, b_i):
    for i in range(len(bif)):
        leeen = np.unique(np.array(bif[i]).round(3))
        if len(leeen)==2**b_i:
            #print('The value of r at bifurcation 1:' , r_range[i])
            break
    return r_range[i]
```
