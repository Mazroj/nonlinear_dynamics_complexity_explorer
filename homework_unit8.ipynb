```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dat = pd.read_csv("amplitude.dat",header= None)[0]
dat = np.array(dat)
dat
```

```python
def rearange(dat,m,t):
    hist = []
    for i in range(len(dat)-(m-1)*t):
        dd = np.zeros(m)
        for j in range(m):
            dd[j] = dat[i+j*t]
        #print(dd)
        hist.append(dd)
    return(hist)


```

```python
data = rearange(dat,7,8)
data = np.array(data)
data
```

```python
plto = data[:,:2].transpose()
```
